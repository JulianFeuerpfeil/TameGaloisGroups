# Examples for the paper "Tame Galois Groups, Linking Numbers and Mildness"
by J. Feuerpfeil, O. Hamza, and D. Lim


We provide the concrete linking data for the Examples considered in Section 5. It requires the functionalites from [OSCAR](https://docs.oscar-system.org/) and the additional file "general_linking_diagrams.jl", which can be found too in the GitHub. 

The following code has to be evaluated first to load the relevant functionalities. The different examples can be evaluated independently. Note that the first evaluation might take longer, since the relevant modules have to be loaded first.

```julia
# Load Oscar and the necessary file containing the code to create the linking diagrams

using Oscar
include("general_linking_diagrams.jl")

Qt,t=polynomial_ring(QQ,"t")
empty_list=AbsSimpleNumFieldOrderIdeal[]
```

## Section 5.1 "Examples Theorem A"

```julia
# Example 5.1.1

# Set up the field and S_0
Q,_=rationals_as_number_field()
Z=maximal_order(Q)
S0=[7*Z,13*Z,19*Z,37*Z]

# define S1
S1=[10639*Z, 826093*Z]


create_linking_diagram_of_K(Q,3, collect(union(S0,S1)),empty_list, empty_list)
```

```julia
# Example 5.1.2 (a)
K,a=number_field(t^2-3)
OK=maximal_order(K)
a=(OK)(a)
S0=AbsSimpleNumFieldOrderIdeal[7*OK,ideal(OK,-a+4)]

# create Koch set and assert that it is a Koch set
SS=AbsSimpleNumFieldOrderIdeal[5*OK]
@show is_trivial(generalized_class_group(SS,empty_list;n_quo=3))

# define S1
S1=AbsSimpleNumFieldOrderIdeal[ideal(OK,1063), ideal(OK,593), ideal(OK,531*a+2224)]

# create the linking diagram
create_linking_diagram_of_K(K,3, collect(union(S0,S1)),SS,empty_list)
```

```julia
# Example 5.1.2 (b)
K,a=number_field(t^2-79)
OK=maximal_order(K)
a=(OK)(a)
S0=AbsSimpleNumFieldOrderIdeal[ideal(OK,7,a+4),ideal(OK,13,a+1)]

# create Koch set and assert that it is a Koch set
SS=AbsSimpleNumFieldOrderIdeal[ideal(OK,7,a+3),ideal(OK,11)]
T=AbsSimpleNumFieldOrderIdeal[ideal(OK,3,a+1)]
#check that SS is a Koch set
@show is_trivial(generalized_class_group(SS,T;n_quo=3))

# define S1
S1=AbsSimpleNumFieldOrderIdeal[ideal(OK,65*a+66),ideal(OK,17389,a+4273),ideal(OK,-2919*a+26594),ideal(OK,86121199, a + 5709782)]

# create the linking diagram
create_linking_diagram_of_K(K,3, collect(union(S0,S1)),SS,T)
```

## Section 5.2 "Examples for 1-RAAGs"

```julia
# Example 5.2.1

# Set up the field and S_0
Q,_=rationals_as_number_field()
Z=maximal_order(Q)

# define S
S=[7*Z,13*Z,181*Z,5563*Z]

create_linking_diagram_of_K(Q,3,S,empty_list, empty_list)
```

```julia
# Example 5.2.2 

# Set up number field
K,a=number_field(t^2+31)
OK=maximal_order(K)
b=a//2+3//2
a=(OK)(a)

# create and verify Koch set
SS=AbsSimpleNumFieldOrderIdeal[ideal(OK,7,a+2)]
T=[ideal(OK,2,(OK)(b))]
@show is_trivial(generalized_class_group(SS,T;n_quo=3))

@show SS,T
# define S
S=AbsSimpleNumFieldOrderIdeal[
ideal(OK,7, a + 5),
ideal(OK,13),
ideal(OK,4129),
ideal(OK,3613),
ideal(OK,257443, a + 245436),
ideal(OK,1062643, a + 778479)
 ]

create_linking_diagram_of_K(K,3,S,SS,T)
```

# Section 5.3 "Examples which are not $(\mathbf{X},e)$-RAAG

Note that we don't have an influence on the internal choice of generator, thus the linking numbers might not appear quite as they are claimed in the paper. It is always possible to multiply all values in a row/ column by an element $\mathbb{F}_p^\times$.

Since we are always working with $p=3$, the symbol $2$ can be read as $-1$.

```julia
# Example 5.3 (a)

# set up field
Q,_=rationals_as_number_field()
Z=maximal_order(Q)

# define S
S=[31*Z,163*Z,409*Z,433*Z]

create_linking_diagram_of_K(Q,3,S,empty_list, empty_list)
```

```julia
# Example 5.3 (b)

# set up field
K,a=number_field(t^2-5)
OK=maximal_order(K)
b=(OK)(a//2+1//2)
a=(OK)(a)


# define Koch tuple
SS=AbsSimpleNumFieldOrderIdeal[ideal(OK,2)]
@show is_trivial(generalized_class_group(SS,empty_list;n_quo=3))


# define S
S=AbsSimpleNumFieldOrderIdeal[ideal(OK,283), ideal(OK,4*a-1),ideal(OK,353), ideal(OK,-b-94),ideal(OK,-343*b+163)]

create_linking_diagram_of_K(K,3,S,SS,empty_list)
```
